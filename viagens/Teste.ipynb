```python
# l = ['nomeorgaosuperior', 'nomeorgaosolicitante', 'companhiaaerea', 'noshow', 'remarcado', 'situacaobilhete']
l = ['situacaobilhete']
for i in range(0, len(l)):
    v = viagens[l[i]]
    v = v.to_list()
    v1=[]
    
    for j in range(0, len(v)):
        a = str(v[j])
        v1.append(a.strip())

    v2 = np.unique(v1)
    s = l[i]+'.txt'
    f = open(s, 'w')
    for i in range(0, len(v2)):
        f.write(str(v2[i]) + ';')
```
